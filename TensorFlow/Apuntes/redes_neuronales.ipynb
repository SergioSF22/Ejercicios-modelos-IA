## **¿Qué es una Red Neuronal?**

Una **red neuronal** es un modelo matemático inspirado en el **cerebro humano**. Está compuesta por nodos llamados **neuronas**, que están organizados en **capas**. Estas redes se entrenan para **aprender patrones** en los datos a través de un proceso llamado **propagación hacia adelante** y **retropropagación**.
| Año        | Hito                               | Descripción                                                                           |
| ---------- | ---------------------------------- | ------------------------------------------------------------------------------------- |
| 1943       | McCulloch & Pitts                  | Primera neurona artificial. Muy básica.                                               |
| 1958       | Rosenblatt - Perceptrón            | Modelo capaz de aprender. Solo resolvía problemas lineales.                           |
| 1970s      | “Invierno de la IA”                | Falta de resultados prácticos. Se pierde interés.                                     |
| 1986       | Retropropagación (Backpropagation) | Permite entrenar redes con muchas capas. Revive el interés.                           |
| 1990s      | RNN y CNN                          | Arquitecturas más complejas aparecen. Problemas de hardware limitan su uso.           |
| 2012       | AlexNet                            | Gana el concurso ImageNet. Inicia el auge del Deep Learning.                          |
| Actualidad | Transformers, GPT, BERT            | Modelos de gran escala que revolucionan el procesamiento de lenguaje, imágenes y más. |

Los componentes principales de una Red Neuronal son:

1. **Neurona (Nodo)**: Cada **neurona** recibe entradas, las procesa, y luego envía una salida a las neuronas de la siguiente capa. La salida de una neurona se calcula aplicando una **función de activación**.
2. **Capas:** Las neuronas se agrupan en **capas**. Cada capa realiza una operación sobre los datos que le llegan de la capa anterior. Existen varios tipos de capas según la tarea:
    - **Capa de entrada (Input Layer)**: Es la capa inicial donde los datos entran en la red.
    - **Capas ocultas (Hidden Layers)**: Son las capas intermedias que procesan los datos.
    - **Capa de salida (Output Layer)**: La capa final que genera las predicciones o resultados.
3. **Conexiones (Pesos)**: Las neuronas están conectadas entre sí, y cada conexión tiene un **peso** que determina la importancia de esa conexión. Los pesos son ajustados durante el entrenamiento para mejorar las predicciones del modelo.
    
    ¿Cómo funciona una Red Neuronal?
    
    1. **Propagación hacia adelante**:
        - **Entrada**: Los datos de entrada se pasan a través de las neuronas de la capa de entrada.
        - **Cálculo**: Cada neurona en las capas ocultas toma las salidas de las neuronas anteriores, las multiplica por los pesos, las pasa por una **función de activación** y produce una salida.
        - **Salida**: La última capa produce la salida final de la red (por ejemplo, una clasificación, un valor numérico, etc.).
    2. **Entrenamiento**: Durante el entrenamiento, las redes neuronales ajustan los **pesos** para minimizar el error en las predicciones. Esto se hace a través del algoritmo de **retropropagación** y el **descenso de gradiente**.

    

## ¿Qué es una Capa en una Red Neuronal?

Una **capa** es un conjunto de **neuronas** que realizan una operación en los datos que fluyen a través de ellas. Las capas tienen un papel específico en el modelo, y dependiendo del tipo de red y tarea, pueden ser diferentes:

1. **Capa densa (Fully Connected Layer)**:
    - Es la capa más común en redes neuronales tradicionales.
    - Cada neurona está conectada a todas las neuronas de la capa anterior y siguiente.
    - Se utiliza principalmente en redes neuronales profundas y modelos de clasificación o regresión.

```python
tf.keras.layers.Dense(units=64, activation='relu')
```

2. **Capa de Convolución (Conv2D)**:
    - Se usa en redes neuronales convolucionales (CNN), especialmente para procesamiento de imágenes.
    - Se encarga de extraer características locales (como bordes, texturas) en imágenes.
    - Es una capa especializada en el procesamiento espacial de datos.

```python
tf.keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu')
```

3. **Capa de Pooling (MaxPooling)**:
    - Se usa en redes convolucionales para reducir el tamaño de la salida y preservar las características importantes.
    - Hace un "muestreo" de las características detectadas por las capas de convolución.

```python
tf.keras.layers.MaxPooling2D(pool_size=(2, 2))
```

4. **Capa Recurrente (RNN, LSTM, GRU)**:
    - Se utiliza en redes neuronales recurrentes para procesar secuencias de datos (como texto o series temporales).
    - Mantiene una memoria de los datos previos, lo que es útil para tareas como traducción de idiomas o análisis de sentimientos.

```python
tf.keras.layers.LSTM(64)
```

Proceso de Propagación hacia Adelante en una Capa:

1. Cada neurona de la capa recibe una entrada desde la capa anterior.
2. Calcula una ponderación de estas entradas (multiplicando cada entrada por su peso).
3. Aplica una función de activación (como  `ReLU`,  `sigmoid`,  `tanh`, etc.) al resultado para decidir si esa neurona se activa o no.
4. La salida de la neurona se pasa a las neuronas de la capa siguiente.

### FUNCIONES DE ACTIVACIÓN
| Función    | Uso típico             | Ideal para...                              |
| ---------- | ---------------------- | ------------------------------------------ |
| Sigmoid    | Salida binaria (0 o 1) | Clasificación binaria                      |
| Tanh       | Capas ocultas          | Cuando necesitas centrado en cero          |
| ReLU       | Capas ocultas          | Deep learning, visión por computadora      |
| Leaky ReLU | Capas ocultas          | Evitar neuronas muertas                    |
| Softmax    | Capa de salida         | Clasificación multiclase (más de 2 clases) |

```python
import numpy as np

# Definimos la función sigmoide, que devuelve valores entre 0 y 1
def sigmoid(x):
    return 1 / (1 + np.exp(-x))  # Fórmula de la función sigmoide

# Definimos la función tanh (hiperbólica tangente), que devuelve valores entre -1 y 1
def tanh(x):
    return np.tanh(x)  # NumPy ya tiene esta función implementada

# Definimos la función ReLU (Rectified Linear Unit)
def relu(x):
    return np.maximum(0, x)  # ReLU devuelve x si es positivo, si no, devuelve 0

# Definimos la función Leaky ReLU, una variante de ReLU que permite un pequeño valor negativo cuando x < 0
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)  # Si x > 0 devuelve x, si no, devuelve alpha * x

# Creamos un arreglo de 100 valores equidistantes entre -10 y 10
x = np.linspace(-10, 10, 100)

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(x, sigmoid(x), label="Sigmoid")
plt.plot(x, tanh(x), label="Tanh")
plt.plot(x, relu(x), label="ReLU")
plt.plot(x, leaky_relu(x), label="Leaky ReLU")
plt.title("Funciones de activación")
plt.legend()
plt.grid(True)
plt.show()
```
