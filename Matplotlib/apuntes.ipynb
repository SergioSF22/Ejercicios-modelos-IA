# Matplotlib
Es una librería para el trazado de gráficos en `2D` y `3D` empleada en el mundo del data science y machine learning.
-   EDA: Análisis Exploratorio de datos

```python
# Importación de la librería
import matplotlib.pyplot as plt
import seaborn as sns
```

### Gráficos de líneas

Representación gráfica de datos númericos mediante lineas.

```python
# Definimos un conjunto de datos
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

#---------------CREACIÓN SIMPLE----------------
# Creamos el gráfico de líneas
plt.plot(x, y, color='red', linestyle="solid")

# Añadimos un título y etiquetas
plt.title("Gráfica de líneas")
plt.xlabel("x")
plt.ylabel("y")

# Mostrar el gráfico
plt.show()

#---------------CREACIÓN MEDIANTE POO----------------
# Preparamos el área del gráfico (fig) y el gráfico en sí (ax) utilizando plt.subplots()
fig, ax = plt.subplots()

# Añadimos los datos al gráfico
ax.plot(x, y)

# Modificamos los parámetros del gráfico
ax.set(title="Gráfico de líneas",
        xlabel="x",
        ylabel="y")
```

### Gráficos de barras

```python
# Definimos un conjunto de datos
x = [2, 1, 5, 4]
y = [1, 4, 16, 25]

#---------------CREACIÓN SIMPLE----------------
# Creamos el gráfico de líneas
plt.bar(x, y, color='red')

# Añadimos un título y etiquetas
plt.title("Gráfico de líneas")
plt.xlabel("x")
plt.ylabel("y")

# Mostrar el gráfico
plt.show()

#---------------CREACIÓN MEDIANTE POO----------------
# Preparamos el área del gráfico (fig) y el gráfico en sí (ax) utilizando plt.subplots()
fig, ax = plt.subplots()

# Añadimos los datos al gráfico
ax.bar(x, y)

# Modificamos los parámetros del gráfico
ax.set(title="Gráfico de líneas",
        xlabel="x",
        ylabel="y")
```

### Histogramas
Representa la frecuencia de aparición de los valores

```python
# Definimos un conjunto de datos
x = [1, 2, 2, 2, 3, 3, 4, 4, 4, 4, 5, 5, 6, 6, 6, 6, 7, 8, 9, 9]

#---------------CREACIÓN SIMPLE----------------
# Creamos el histograma
plt.hist(x, color='orange', edgecolor="black")

# Añadimos un título y etiquetas
plt.title("Gráfica de líneas")
plt.xlabel("x")
plt.ylabel("y")

# Mostrar el gráfico
plt.show()

#---------------CREACIÓN MEDIANTE POO----------------
# Preparamos el área del gráfico (fig) y el gráfico en sí (ax) utilizando plt.subplots()
fig, ax = plt.subplots()

# Añadimos los datos al gráfico
ax.hist(x)

# Modificamos los parámetros del gráfico
ax.set(title="Gráfico de líneas",
        xlabel="x",
        ylabel="y")
```

### GRÁFICO PASTEL

Representa datos categóricos en los que cada sector va a representar una categoría dentro de mi conjunto de datos.

```python
# Definimos un conjunto de datos
tamanos = [10, 20, 30, 40, 50]
valores = [1, 2, 3, 4, 5]

#---------------CREACIÓN SIMPLE----------------
# Crear gráfico de pastel
plt.figure(figsize=(6, 6))  # Tamaño del gráfico
plt.pie(tamanos, labels=valores, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Título
plt.title("Gráfico pastel")

# Mostrar gráfico
plt.show()

#---------------CREACIÓN MEDIANTE POO----------------
# Preparamos el área del gráfico (fig) y el gráfico en sí (ax) utilizando plt.subplots()
fig, ax = plt.subplots()

# Añadimos los datos al gráfico
ax.pie(tamanos, labels=valores, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

# Modificamos los parámetros del gráfico
ax.set(title="Gráfico pastel")
```

### Gráfico de densidad

Muestran la distribución de un conjunto de datos NUMÉRICOS

```python
import numpy as np

# Generar un conjunto de datos
datos = np.random.normal(loc=0, scale=2, size=1000)

# Crear un gráfico de densidad
sns.kdeplot(datos, color='red')

# Añadimos un título y etiquetas
plt.title("Gráfica de densisas")
plt.xlabel("x")
plt.ylabel("y")

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import statsmodels.api as sm
plt.style.use('seaborn-v0_8-whitegrid')

# Simulo tu DataFrame con distribución parecida para ejemplo
np.random.seed(0)
datos = np.concatenate([np.random.randint(1, 20, 18000), np.random.randint(20, 500, 600), [91079]])
df = pd.DataFrame({'num_personas': datos})

# Crear figura y ejes
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Histograma con escala log
sns.histplot(df['num_personas'], bins=50, ax=axs[0, 0])
axs[0, 0].set_xscale('log')
axs[0, 0].set_title('Histograma (escala log)')
axs[0, 0].set_xlabel('Número de personas por grupo')
axs[0, 0].set_ylabel('Cantidad de grupos')

# Boxplot limitado a 0-50
sns.boxplot(x=df['num_personas'], ax=axs[0, 1])
axs[0, 1].set_xlim(0, 50)
axs[0, 1].set_title('Boxplot (limite a 0–50)')
axs[0, 1].set_xlabel('Número de personas por grupo')

# Violinplot limitado a 0-50
sns.violinplot(x=df[df['num_personas'] < 50]['num_personas'], ax=axs[1, 0])
axs[1, 0].set_title('Violinplot (<50)')
axs[1, 0].set_xlabel('Número de personas por grupo')

# ECDF con escala log
ecdf = sm.distributions.ECDF(df['num_personas'])
axs[1, 1].plot(ecdf.x, ecdf.y)
axs[1, 1].set_xscale('log')
axs[1, 1].set_title('ECDF (escala log)')
axs[1, 1].set_xlabel('Número de personas por grupo')
axs[1, 1].set_ylabel('Proporción acumulada')

# Título general
fig.suptitle('Distribución de personas por grupo - Comparativa de métodos', fontsize=18)

# Ajuste de layout para dejar espacio al título
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```
