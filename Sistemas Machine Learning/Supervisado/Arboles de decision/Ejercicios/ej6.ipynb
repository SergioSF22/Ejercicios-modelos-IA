## 6. Árbol de decisión con digits (Ejercicio muestra)

```python
from sklearn.datasets import load_digits
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# Cargamos el dataset y definimos las variables predictoras y objetivo
digits = load_digits()
X = digits.data 
y = digits.target

# Dividimos los datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Creamos el modelo de arbol de clasificación
model = DecisionTreeClassifier(max_depth=10, random_state=42)

# Entrenamos el modelo
model.fit(X_train, y_train)

# Creamos las predicciones
y_pred = model.predict(X_test)

# Evaluamos el modelo con Accuracy
acurracy = accuracy_score(y_test, y_pred)
print(f"Precisión del modelo: {acurracy * 100:.2f}%")

# Visualizar algunas imágenes con su predicción
fig, axes = plt.subplots(2, 5, figsize=(10, 5)) # 2 filas con 5 figuras de 10x5 pulgadas
for i, ax in enumerate(axes.flat): #axes.flat aplana la matriz de 2x5 a 1x10 para iterar mejor sobre ella.
    ax.imshow(X_test[i].reshape(8, 8), cmap="gray")
    ax.set_title(f'Pred: {y_pred[i]}')
    ax.axis("off")
plt.show()

# Visualizar el árbol de decisión
plt.figure(figsize=(20, 10))
tree.plot_tree(model, filled=True, feature_names=[f'Pixel {i}' for i in range(X.shape[1])], class_names=[str(i) for i in range(10)])
plt.show()
```
