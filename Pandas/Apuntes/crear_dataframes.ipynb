### ¿Qué diferencia hay entre un CSV, un Dataset y un Dataframe?


```python
# Leer el CSV y convertirlo en un DataFrame
df = pd.read_csv("Product_v6.csv")
print("Todas las columnas:", df.columns)  # Mostrar todas las columnas

# Seleccionar una columna (ejemplo: 'value') y convertirla en un array de NumPy
array_np = np.array(df["value"])

# Aplicar una operación con NumPy (ejemplo: raíz cuadrada)
resultado_np = np.sqrt(array_np)

# Añadir los resultados como una nueva columna en el DataFrame
df["raiz_cuadrada"] = resultado_np

# Guardar el DataFrame actualizado en un nuevo CSV (opcional)
df.to_csv("datos_modificados.csv", index=False)
```

### Crear columnas con valores aleatorios

Módulo random de Python
| Función | Descripción | Parámetros | Ejemplo |
|---------|------------|------------|---------|
| `random.random()` | Genera un número flotante aleatorio entre 0 y 1. | *Sin parámetros.* | `random.random() → 0.74` |
| `random.uniform(a, b)` | Genera un número flotante aleatorio en el rango [a, b]. | `a`: mínimo, `b`: máximo | `random.uniform(10, 20) → 15.32` |
| `random.randint(a, b)` | Genera un número entero aleatorio en el rango [a, b] (incluye ambos extremos). | `a`: mínimo, `b`: máximo | `random.randint(1, 100) → 57` |
| `random.randrange(start, stop, step)` | Similar a `randint`, pero con un paso definido. | `start`: inicio, `stop`: fin, `step`: incremento | `random.randrange(0, 100, 5) → 35` |
| `random.choice(seq)` | Devuelve un elemento aleatorio de una lista, tupla o string. | `seq`: iterable | `random.choice(['rojo', 'azul', 'verde']) → 'azul'` |
| `random.choices(seq, k=n)` | Devuelve una lista con `n` elementos aleatorios (con reemplazo). | `seq`: iterable, `k`: cantidad | `random.choices(['A', 'B', 'C'], k=3) → ['C', 'A', 'B']` |
| `random.sample(seq, k=n)` | Devuelve una lista con `n` elementos aleatorios sin repetición. | `seq`: iterable, `k`: cantidad | `random.sample(range(1, 10), k=3) → [2, 7, 9]` |
| `random.shuffle(seq)` | Mezcla los elementos de una lista *in situ*. | `seq`: lista modificable | `lista = [1,2,3]; random.shuffle(lista) → [3,1,2]` |
| `random.gauss(mu, sigma)` | Genera un número con distribución normal (gaussiana). | `mu`: media, `sigma`: desviación estándar | `random.gauss(50, 10) → 53.8` |
| `random.betavariate(alpha, beta)` | Genera un número con distribución Beta. | `alpha`, `beta`: parámetros de la Beta | `random.betavariate(2, 5) → 0.26` |
| `random.expovariate(lambd)` | Genera un número con distribución exponencial. | `lambd`: λ (inverso de la media) | `random.expovariate(1.5) → 0.8` |


np.random de Numpy
| Función | Descripción | Parámetros | Ejemplo |
|---------|------------|------------|---------|
| `np.random.rand(n)` | Genera `n` números flotantes entre **0 y 1**. | `n`: cantidad | `np.random.rand(3) → [0.21, 0.78, 0.44]` |
| `np.random.randn(n)` | Genera `n` números flotantes con distribución normal estándar (**media=0, sigma=1**). | `n`: cantidad | `np.random.randn(3) → [-1.2, 0.4, 1.7]` |
| `np.random.randint(a, b, size=n)` | Genera `n` enteros en el rango `[a, b)`. | `a`: mínimo, `b`: máximo, `size`: cantidad | `np.random.randint(1, 10, size=3) → [3, 8, 5]` |
| `np.random.choice(seq, size=n, replace=True)` | Toma `n` elementos de `seq`, con o sin repetición. | `seq`: iterable, `size`: cantidad, `replace`: permitir repetición | `np.random.choice([1,2,3], size=2, replace=False) → [2,1]` |
| `np.random.permutation(seq)` | Devuelve una **lista aleatorizada** de los elementos de `seq`. | `seq`: iterable | `np.random.permutation([1,2,3]) → [3,1,2]` |
| `np.random.normal(mu, sigma, size=n)` | Genera `n` valores con distribución normal. | `mu`: media, `sigma`: desviación estándar, `size`: cantidad | `np.random.normal(50, 10, size=3) → [45.8, 51.3, 60.2]` |
